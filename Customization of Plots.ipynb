```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
titanic.info()
```

```python
import matplotlib.pyplot as plt
```

```python
titanic.age.plot(figsize=(12,8))
plt.show()
```

```python
titanic.age.plot(figsize=(12,8), fontsize = 15)
plt.show()
```

```python
titanic.age.plot(figsize=(12,8), fontsize = 15, c = "red")
plt.show()
```

```python
titanic.age.plot(figsize=(12,8), fontsize = 15, c = "red", linestyle = "--")
plt.show()
```

```python
titanic.age.plot(figsize=(12,8), fontsize = 15, c = "red", linestyle = ":")
plt.show()
```

```python
titanic.age.plot(figsize=(12,8), fontsize = 15, c = "red", linestyle = "-")
plt.show()
```

```python
titanic.age.plot(figsize=(12,8), fontsize = 15, c = "red", linestyle = "-")
plt.title("Ttianic - Ages", fontsize = 15)
plt.show()
```

```python
titanic.age.plot(figsize=(12,8), fontsize = 15, c = "red", linestyle = "-")
plt.title("Ttianic - Ages", fontsize = 15)
plt.legend(loc = "best", fontsize=15)
plt.show()
```

```python
titanic.age.plot(figsize=(12,8), fontsize = 15, c = "red", linestyle = "-")
plt.title("Ttianic - Ages", fontsize = 15)
plt.legend(loc = "best", fontsize=15)
plt.xlabel("Passenger No", fontsize = 13)
plt.ylabel("Age", fontsize=13)
plt.grid()
plt.show()
```

```python
titanic.age.plot(figsize=(12,8), fontsize = 13, c = "red", linestyle = "-", xlim = (200,400))
plt.title("Ttianic - Ages", fontsize = 15)
plt.legend(loc = 3, fontsize=15)
plt.xlabel("Passenger No", fontsize = 13)
plt.ylabel("Age", fontsize=13)
plt.grid()
plt.show()
```

```python
titanic.age.plot(figsize=(12,8), fontsize = 13, c = "red", linestyle = "-", xlim = (200,400), ylim = (20,40))
plt.title("Ttianic - Ages", fontsize = 15)
plt.legend(loc = 3, fontsize=15)
plt.xlabel("Passenger No", fontsize = 13)
plt.ylabel("Age", fontsize=13)
plt.grid()
plt.show()
```

```python
xticks = [x for x in range(0,901,50)]
```

```python
yticks = [y for y in range(0,81,5)]
```

```python
titanic.age.plot(figsize=(12,8), fontsize = 13, c = "red", linestyle = "-", xlim = (0,900), ylim = (0,80),
                xticks = xticks, yticks = yticks)
plt.title("Ttianic - Ages", fontsize = 15)
plt.legend(loc = 3, fontsize=15)
plt.xlabel("Passenger No", fontsize = 13)
plt.ylabel("Age", fontsize=13)
plt.grid()
plt.show()
```

```python
titanic.age.plot(figsize=(12,8), fontsize = 13, c = "red", linestyle = "-", xlim = (0,900), ylim = (0,80),
                xticks = xticks, yticks = yticks, rot=45)
plt.title("Ttianic - Ages", fontsize = 15)
plt.legend(loc = 3, fontsize=15)
plt.xlabel("Passenger No", fontsize = 13)
plt.ylabel("Age", fontsize=13)
plt.grid()
plt.show()
```
