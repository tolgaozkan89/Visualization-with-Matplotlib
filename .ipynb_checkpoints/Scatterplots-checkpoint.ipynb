```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("seaborn")
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
titanic.plot(kind = "scatter", figsize = (15,8), x="age", y="fare")
plt.show()
```

```python
titanic.plot(kind = "scatter", figsize = (15,8), x="age", y="fare", c = "red")
plt.show()
```

```python
titanic.plot(kind = "scatter", figsize = (15,8), x="age", y="fare", c = "red", marker = "x")
plt.show()
```

```python
titanic.plot(kind = "scatter", figsize = (15,8), x="age", y="fare", c = "red", marker = "x", s = 50)
plt.show()
```

```python
titanic.plot(kind = "scatter", figsize = (15,8), x="age", y="fare", c = "survived", marker = "x", s = 50)
plt.show()
```

```python
titanic.plot(kind = "scatter", figsize = (15,8), x="age", y="fare", c = "survived", marker = "x", s = 50, colormap = "viridis")
plt.show()
```

```python
titanic.plot(kind = "scatter", figsize = (15,8), x="age", y="fare", c = "pclass", marker = "x", s = 50, colormap = "viridis")
plt.show()
```
