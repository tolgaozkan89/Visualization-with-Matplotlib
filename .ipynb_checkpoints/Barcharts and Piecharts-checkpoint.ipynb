```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("seaborn")
```

```python
summer_2012 = pd.read_csv("summer_2012.csv", index_col = "Country")
```

```python
summer_2012
```

```python
summer_2012.head()
```

```python
summer_2012.Medal.plot(kind = "line", figsize = (12,8), fontsize = 12)
plt.show()
```

```python
summer_2012.Medal.plot(kind = "bar", figsize = (12,8), fontsize = 12)
plt.show()
```

```python
summer_2012.Medal.plot(kind = "barh", figsize = (12,8), fontsize = 12)
plt.show()
```

```python
summer_2012.Medal.plot(kind = "pie", figsize = (12,8), fontsize = 12)
plt.show()
```
